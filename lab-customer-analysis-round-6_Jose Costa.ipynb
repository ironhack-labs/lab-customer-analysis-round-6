```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

data=pd.read_csv("files_for_lab\csv_files\marketing_customer_analysis.csv")
data
```

```python
numeric = data.select_dtypes(include=[np.number])
numeric
```

```python
numeric_x=numeric.drop(["Number of Open Complaints","Number of Policies"],axis=1)
numeric_x["Income"]=numeric_x["Income"].replace(0,numeric_x["Income"].mean())
numeric_x
```

# One Hot/Label Encoding (categorical).

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

categ = data.select_dtypes(include=[np.object])
categ
```

```python
categ_x=categ.drop(["Customer","Effective To Date"],axis=1)
categ_x
```

```python
for col in categ_x.columns:
    categ_x[col]=LabelEncoder().fit(categ_x[col]).transform(categ_x[col])

categ_x
```

# Concat DataFrames

```python
numeric_x
```

```python
data_w=pd.DataFrame(np.concatenate((numeric_x,categ_x),axis=1)).reset_index(drop=True)
data_w
```

```python
data_w.columns = list(numeric_x.columns) + list(categ_x.columns)
data_w
```

# Train-test split.

```python
Y=data_w["Total Claim Amount"].reset_index(drop=True)
X=data_w.drop(["Total Claim Amount"],axis=1).reset_index(drop=True)
Y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.4,random_state=100)
```

```python
import statsmodels.api as sm

model=sm.OLS(y_train,x_train).fit()

X.to_csv("final_X.csv")
Y.to_csv("final_Y.csv")
```

```python
predictions=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

#R2

r2=r2_score(y_test,predictions)
r2
```

# Apply linear regression.

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(x_train, y_train)
```

```python
predictions=reg.predict(x_test)

#R2

r2=r2_score(y_test,predictions)
r2
```

### MSE

```python
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y_test,predictions)
mse
```

### RMSE

```python
rmse=mean_squared_error(y_test,predictions,squared=False)
rmse
```

### MAE

```python
from sklearn.metrics import mean_absolute_error

mae=mean_absolute_error(y_test,predictions)
mae
```
