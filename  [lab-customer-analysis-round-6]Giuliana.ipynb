```python
# One Hot/Label Encoding (categorical)

# I decided to drop the customers because it was't helping the model, since it added litteraly thousands of columns with not much valuable information 

hot_categoricals = pd.get_dummies(categoricals.drop(['customer'], axis = 1), drop_first = True)
hot_categoricals
```

```python
concat2 = pd.merge(numerical, hot_categoricals, left_index=True, right_index=True)
concat2
```

```python
# Linear Regression 

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

Y = concat2['total_claim_amount']
X = concat2.drop(['total_claim_amount'], axis=1) 

lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
model.score(X,Y)
```

```python
# Train-test split

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)



```

```python
# Apply linear regression

Y = Y_train
X = X_train

lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
model.score(X,Y)

predictions = lm.predict(X_test) 
predictions


print("R2 value is = ", round(r2_score(Y_test, predictions), 2))
print("The intercept of the model is = ", lm.intercept_)
print("The coefficients of the models are: ", lm.coef_)
```

```python
# Model Validation Description: R2. MSE. RMSE. MAE

mse = mean_squared_error(Y_test, predictions)
print("The Mean Squared error is: ", mse)
sqrt = np.sqrt(mse)
print("The root mse of the model is: ", sqrt)
mae = mean_absolute_error(Y_test, predictions)
print("The Mean absolute error is:", mae)
```
