```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

#Libary for predictions
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# These Libs are for stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
#1. Get the data:
df = pd.read_csv(r'C:\Users\ljant\Desktop\Ironhack\06_Advanced_Statistics\lab-customer-analysis-round-6\files_for_lab\csv_files\marketing_customer_analysis.csv')
df
```

```python
#2. Organising/dealing with the df:

def header_function (data):
    cols = []
    for i in range (len(data.columns)):
        cols.append(data.columns[i].lower().replace(' ','_'))
    
    data.columns = cols

    return data

header_function(df)
```

```python
# 3. Processing the data
# a) xy-split of the data

df.shape #(9134, 24)

from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.2, random_state=11)
df_train #(7307, 24)
```

```python
''' 
I'm not going to One Hot/Label Encoding every categorical column, to keept the table manageable
I'm choosing:
- 'state','response','coverage', 'location_code','policy_type', 'sales_channel'
'''

df_cat = pd.get_dummies(df_cat[['state','response','coverage', 'location_code','policy_type', 'sales_channel']], dtype = int)
df_cat
#[7307, 20]
```

```python
# Adjusting column headers

def header_function (data):
    cols = []
    for i in range (len(data.columns)):
        cols.append(data.columns[i].lower().replace(' ','_'))
    
    data.columns = cols

    return data

header_function(df_cat)
```

```python
# d) concat dataframes

df_model = pd.concat([df_cat, normalised_data], axis = 1).reset_index(drop = True)
df_model
```

```python
# 4. Linear regression

#a) Build regression with 'total_claim_amount' as target value; testing with income

Y = target
X = df_model['income']

# The model: Y = inter + coef. * X

X = sm.add_constant(X) # here adding the intercept

# The model:
model = sm.OLS(Y, X).fit()
model.summary()

```

```python
Y = target
X = df_model['monthly_premium_auto']

# The model: Y = inter + coef. * X

X = sm.add_constant(X) # here adding the intercept

# The model:
model = sm.OLS(Y, X).fit()
model.summary()
```

```python
'''
Model tests the relationshop between 'total_claim_amount' and 'income' --> does the 'total_claim'amount' depends on 'income'?
--> initial hypothesis: "Customers with higher income have a lower total claim amount than customers with lower income."

Results interpretation:
- Negative coef --> As 'income' increases, the total_claim_amount decreases --> in accordance with the hypothesis
- Postive coef --> As 'monthly_premium_auto' increases the total_claim_amount increases as well
- R² = 0.126 
--> 'income' is not a great variable (R² < 0.5) to explain the variation in total_claim_amount, but still a good approach. 
    In contrast 'monthly_premium_auto' explains 39.3 % of the variance in total_claim_amount.
- F-Statistcs suggest that there is a linear relationship between 'income' and 'total_claim_amount'
- T-test in multiple regression suggests that both variables ('income' and 'monthly_premium_auto') are relevant for the 
  regression model.

'''
```

```python
# Multiple regresssion

Y = target
X = df_model

X = sm.add_constant(X)

# The model:
model = sm.OLS(Y, X).fit()
model.summary()
```

```python
# Predictions and error analysis

Y = target
X = df_model['income'].values.reshape(-1, 1) #changes X to 2D format (needed for LinearRegression function)

lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
lm.score(X,Y) # This is the R^2

# This is the same as above
# But now it can be used for predictions:


from sklearn.metrics import mean_absolute_error

predictions = lm.predict(X)

print("R2 value is = ",round(r2_score(Y, predictions),2))
print("The intercept of the model is = ",lm.intercept_)
print("The coefficients of the model are = ",lm.coef_)

mse = mean_squared_error(Y, predictions)
mae = mean_absolute_error(Y, predictions)
print("The mse of the model is = ", round(mse,2))
print("The root mse of the model is = ",round(np.sqrt(mse),2))
print("The mean absolute error of the model is = ",round(mae,2))
```
