```python

import pandas as pd

mark_cust_A = pd.read_excel('marketing_customer_analysis.xlsx')

# fonction processing all the steps to make the data clean for our file
def Data_quarter_Y(mark_cust_A):
    # creation of a list to store columns name after lower and removing spaces
    cols_st = []
    # 
    for i in range(len(mark_cust_A.columns)):
        cols_st.append(mark_cust_A.columns[i].lower().replace(" ","_"))
    mark_cust_A.columns = cols_st
    mark_cust_A = mark_cust_A.rename(columns={"employmentstatus": "employment_status"}).fillna({'response': 'No', 'number_of_open_complaints': 0, 'vehicle_type': 'Other vehicle'})
    mark_cust_A['date'] = pd.to_datetime(mark_cust_A['effective_to_date'])
    mark_cust_A['month'] = mark_cust_A['date'].dt.to_period('M')
    
 
      
    return  mark_cust_A

mark_cust_A = Data_quarter_Y(mark_cust_A)

mark_cust_A
```

```python
# One Hot/Label Encoding (categorical).


Categorical = mark_cust_A.select_dtypes(include='object')
# drop the first Row as it is not relevant for me to have 100 columns about customer
Categorical = Categorical.drop('customer', axis=1)
# chane value for object to 0 and 1
New_Categorical = pd.get_dummies(Categorical)

New_Categorical

```

```python
import pandas as pd
New_df = pd.concat([New_Categorical, Numerical])

New_df = New_df.fillna(0)

New_df

```

```python
# split Data in two segment train(80%) and test (20%)

from sklearn.model_selection import train_test_split

train, test = train_test_split(New_df, test_size=0.2, random_state=42)

print(train, test)
```

```python
# apply linear regression 

Y = train['total_claim_amount']
X = train.drop(['total_claim_amount'], axis=1).fillna(0)
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()

model.summary()
```

```python
#Description:R2.MSE.RMSE.MAE.
import numpy as np

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
lm.score(X,Y)

predictions = lm.predict(X)

print("R2 value is = ",round(r2_score(Y, predictions),2))
print("The intercept of the model is = ",lm.intercept_)
print("The coefficients of the model are = ",lm.coef_)

mse = mean_squared_error(Y, predictions)
mae = mean_absolute_error(Y, predictions)
print("The mse of the model is = ",mse)
print("The root mse of the model is = ",np.sqrt(mse))
print("The mean absolute error of the model is = ",mae)
```
