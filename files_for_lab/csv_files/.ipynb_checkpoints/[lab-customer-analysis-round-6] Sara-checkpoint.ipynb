```python
#Importing libraries and Data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import MinMaxScaler

import statsmodels.api as sm
from statsmodels.formula.api import ols

mc_n = pd.read_csv('datalab4_numerical.csv')
mc_c = pd.read_csv('datalab4_categorical.csv')
```

```python
display(mc_c)
```

```python
mc_c_clean = mc_c.drop(['Unnamed: 0','customer','effective_to_date'], axis = 1)
```

```python
mc_c_clean
```

```python
# One hot enconding

mc_c_clean = pd.get_dummies(mc_c_clean, drop_first=True)
mc_c_clean
```

```python
# Normalizing data 
scaler = MinMaxScaler()
normalized_mc = scaler.fit_transform(mc_n)
normalized_mc = pd.DataFrame(normalized_mc, columns = mc_n.columns)
normalized_mc

# Standardizing columns names

cols2 = []
for i in range(len(normalized_mc.columns)):
    cols2.append(normalized_mc.columns[i].lower().replace(" ","_"))
normalized_mc.columns = cols2
normalized_mc
```

```python
# Concatenating numerical and categorical dataframes

mc_finaldata = pd.concat([normalized_mc, mc_c_clean], axis = 1)
mc_finaldata
```

```python
# Linear Regression
# Train-test split

x = mc_finaldata.drop(['total_claim_amount'], axis =1)
y = mc_finaldata['total_claim_amount']
```

```python
x
```

```python
y
```

```python
# Building a linear model

from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and fit the linear regression model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, Y_train)
```

```python
predictions = lm.predict(X_test)

print("R2 value is = ",round(r2_score(Y_test, predictions),2))
print("The intercept of the model is = ",lm.intercept_)
print("The coefficients of the model are = ",lm.coef_)

mse = mean_squared_error(Y_test, predictions)
mae = mean_absolute_error(Y_test, predictions)
print("The mse of the model is = ",mse)
print("The root mse of the model is = ",np.sqrt(mse))
print("The mean absolute error of the model is = ",mae)
```
