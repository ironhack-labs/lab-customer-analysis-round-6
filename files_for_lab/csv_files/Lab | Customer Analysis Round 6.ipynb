#### Concat DataFrames

```python
categorical = pd.read_csv('merged_clean_ver3_categorical.csv')
categorical.drop(['Unnamed: 0'],axis=1,inplace=True)
categorical.drop(['CUSTOMER'],axis=1,inplace=True)
categorical.drop(['EFFECTIVE TO DATE'],axis=1,inplace=True)
categorical
```

```python
for col in categorical: 
    categorical[col + '_LABEL'] = LabelEncoder().fit(categorical[col]).transform(categorical[col])
```

```python
num_categorical = categorical._get_numeric_data()
```

```python
X = pd.read_csv('X_final_data.csv')
X.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
Y = pd.read_csv('Y_final_data.csv')
Y.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
X
```

```python
num_categorical
```

```python
final_table = pd.DataFrame(np.concatenate((X,num_categorical),axis=1))
final_table.columns = X.columns.append(num_categorical.columns)
```

```python
final_table 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(final_table, Y, test_size=0.4, random_state=100)
```

```python
import statsmodels.api as nm
import statsmodels.api as sm
```

```python
model = sm.OLS(Y_train,X_train).fit()
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test,predictions)
```

```python
# Mean Square error

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y_test, predictions)
print(mse)
```

```python
# root mean squared error

import math

rmse = math.sqrt(mse)
print(rmse)
```
