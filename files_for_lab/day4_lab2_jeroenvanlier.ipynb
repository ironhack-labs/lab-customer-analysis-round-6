```python
import pandas as pd
```

```python
X_normalised_marketing_customer_analysis = pd.read_csv("X_normalised_marketing_customer_analysis.csv",index_col=0)
Y_marketing_customer_analysis = pd.read_csv("Y_marketing_customer_analysis.csv",index_col=0)
X_normalised_marketing_customer_analysis.info()
```

```python
X_cat_normalised_marketing_customer_analysis = X_normalised_marketing_customer_analysis.select_dtypes('object')
X_cat_normalised_marketing_customer_analysis = X_cat_normalised_marketing_customer_analysis.drop(['Customer','Effective To Date'],axis=1)
to_drop_cols = list(X_cat_normalised_marketing_customer_analysis.columns)
X_cat_normalised_marketing_customer_analysis.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first').fit(X_cat_normalised_marketing_customer_analysis)
OHE_objects = encoder.transform(X_cat_normalised_marketing_customer_analysis).toarray()
OHE_objects_names = list(encoder.get_feature_names())

X_completed_data = pd.DataFrame(OHE_objects,columns=OHE_objects_names)
```

```python
cols_to_add = list(X_normalised_marketing_customer_analysis.select_dtypes('float').columns)
#cols_to_add +=['Customer','Effective To Date']
```

```python
for i in cols_to_add:
    X_completed_data[i] = X_normalised_marketing_customer_analysis[i]
```

```python
X_completed_data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_completed_data,
                                                    Y_marketing_customer_analysis,
                                                    test_size=0.3,
                                                    random_state=100)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
model = reg.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

print('r2_score: ',r2_score(y_test,pred))
print('MSE:      ',mean_squared_error(y_test,pred,squared=False))
print('RMSE:     ',mean_squared_error(y_test,pred,squared=True))
print('MAE:      ',mean_absolute_error(y_test,pred))
```
